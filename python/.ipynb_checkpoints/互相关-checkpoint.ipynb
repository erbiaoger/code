### np.correlate 函数

它们的调用参数完全相同。在调用时有三种模式可供选择，它们计算的内容是相同的，但是返回值长度各不相同：

mode = ‘valid’：只返回有效的那一部分相关数据，共$M-N+1$个；

mode = ‘same’：只返回与 等长的那一部分相关数据，共$N$个；

mode = ‘full’：返回全部相关数据，共$M+N-1$个。


```python
np.correlate?
```

```python
import numpy as np
import matplotlib.pyplot as plt

list1 = np.array([np.cos(0.1*i) for i in range(100)])
list2 = np.array([np.cos(0.1*i + np.pi) for i in range(100)])

a = np.correlate(list1, list2, mode='same')

fig, axes = plt.subplots(1,2)
axes[0].plot(list1)
axes[0].plot(list2)

axes[1].plot(a)
```

```python
import numpy as np
import matplotlib.pyplot as plt

list1 = np.array([np.cos(0.1*i) for i in range(100)])
list2 = np.array([np.cos(0.1*i + np.pi) for i in range(200)])

a = np.correlate(list1, list2, mode='valid')

fig, axes = plt.subplots(1,2)
axes[0].plot(list1)
axes[0].plot(list2)

axes[1].plot(a)
```

### scipy.signal.correlate()

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy

list1 = np.array([np.cos(0.1*i) for i in range(100)])
list2 = np.array([np.cos(0.1*i + np.pi) for i in range(100)])

a = scipy.signal.correlate(list1, list2, mode='same')

fig, axes = plt.subplots(1,2)
axes[0].plot(list1)
axes[0].plot(list2)

axes[1].plot(a)
```
