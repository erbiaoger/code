```python
import numpy as np
import scipy.io as io
import glob
import os

dic = []
cwd = os.getcwd()
npyfiles = glob.glob(os.path.join(cwd, '*.npy'))
for file in npyfiles:
    name, _ = file.split('.')
    print(name.split('/')[-1], file)
    data = np.load(file)
    io.savemat(name+'.mat', {name: data})
    dic.append(data)
```
