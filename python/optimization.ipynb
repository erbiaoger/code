```python
# 梯度下降法

def loss_function(X, X_obs):
    return 1/2 * (X - X_obs)**2

def gradient(X, X_obs):
    return X - X_obs

def gradient_descent(X_init, X_obs, learning_rate, num_iterations):
    X = X_init
    loss_old = X
    for i in range(num_iterations):
        grad = gradient(X, X_obs)
        X = X - learning_rate * grad
        
        loss = loss_function(X, X_obs).sum()
        print((loss_old.sum() - loss)/loss)
        loss_old = loss

        print("Iteration %d: loss = %.4f" % (i, loss))
    return X
```

```python
X_init = np.arange(10)/10
X_obs = np.random.randn(10)

X = gradient_descent(X_init, X_obs, 0.1, 30)
print(X_obs, '\n', X)
```

```python
# 牛顿法
import numpy as np

def loss_function(X, X_obs):
    return 1/2 * (X - X_obs)**2

def gradient(X, X_obs):
    return X - X_obs

# 定义二阶导数计算函数
def hessian(X):
    return np.ones(len(X))/len(X)

def gradient_descent(X_init, X_obs, num_iterations):
    X = X_init
    loss_old = X
    for i in range(num_iterations):
        grad = gradient(X, X_obs)
        hess = hessian(X)

        X = X - hess * grad
        
        loss = loss_function(X, X_obs).sum()
        print((loss_old.sum() - loss)/loss)
        loss_old = loss

        print("Iteration %d: loss = %.4f" % (i, loss))
    return X

```
